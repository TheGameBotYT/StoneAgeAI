```python
import numpy as np
import pandas as pd
from StoneAgeGame import StoneAgeGame
from StoneAgeAI import GLIEMonteCarloControl
policy = GLIEMonteCarloControl()

num_episodes=2000000

for i_episode in range(1, num_episodes+1):
    if i_episode % 10000 == 0:
        print('Episode = ', i_episode)
    if i_episode % 1000000 == 0:
        Q = pd.DataFrame.from_dict(policy.Q)
        N = pd.DataFrame.from_dict(policy.N)
        Q.to_pickle('Qrun.p') # Current run = 2.500.000
        N.to_pickle('Nrun.p')
    obj = StoneAgeGame(policy, player_types=['AI', 'AI'])  # env.reset?
    # start_to_play = time.time()
    while not obj.end_of_game():
        obj.play()  # Play epsilon-greedily
    policy.update(obj.states_list, obj.action_list, obj.reward_list, i_episode)
```

```python
Q = pd.read_pickle('Qrun.p')
Q.to_pickle('Q2Mnewstate.p')
N = pd.read_pickle('Nrun.p')
N.to_pickle('N2Mnewstate.p')
```

```python
att = Q.reset_index()
for col in att.columns:
    print(col)
    if type(col) == str:
        sign = 'signed'
    else:
        sign = 'float'
    att[col] = pd.to_numeric(att[col], downcast=sign)
```

```python
index_names = ['Placements', 'Actions', 'SFood', 'OFood',
                  'SelfAgri', 'OppAgri', 'SelfWorkers', 'OppWorkers',
                  'SelfFarm', 'OppFarm', 'SelfHut', 'OppHut',
                  'SelfChop', 'OppChop', 'SelfHunt', 'OppHunt']
att = att.drop(columns='Sum').set_index(index_names)
```
